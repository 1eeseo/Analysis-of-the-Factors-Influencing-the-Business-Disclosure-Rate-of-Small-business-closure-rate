# 열 rename

```python
def rename_columns(df):
    new_columns = []
    for col in df.columns:
        if "분기" in col:
            year, quarter = col.split("년")
            year = year.strip()[-2:]
            quarter = quarter.strip()[0]
            new_columns.append(f"{year}_{quarter}")
        else:
            new_columns.append(col)
    df.columns = new_columns
    return df
```

```python
df2 = rename_columns(df2)
```

```python
print("df2의 변경된 열 이름:\n", df2.columns)
```

# 서울 지역만 추출

```python
def filter_seoul(df):
    filtered_df = df[df['지역'].str.startswith('서울')]
    return filtered_df
```

```python
df2_filtered = filter_seoul(df2)
```
