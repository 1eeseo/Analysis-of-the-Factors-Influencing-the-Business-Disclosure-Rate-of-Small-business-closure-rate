# 요식업, 숙박 및 도소매업 데이터만 남기기


```python
import pandas as pd

# CSV 파일 불러오기
df = pd.read_csv(path1, index_col=0)

# 인덱스 확인 (첫 번째 인덱스는 '날짜'로 되어 있으므로 제외하고 변환)
date_index = df.index[1:]  # 첫 번째 값 '날짜'를 제외
converted_dates = pd.to_datetime(date_index, format='%b.%y').strftime('%Y%m')

# 변환된 날짜를 기존 인덱스에 반영 ('날짜'를 다시 추가)
df.index = ['날짜'] + list(converted_dates)

# 열 이름에서 '소계'를 '소비자물가지수'로 변경
df.columns = df.columns.str.replace('소계', '소비자물가지수')

# 필요한 열 선택
df_filtered = df[['음식 및 숙박','의복·신발']]

# 첫 번째 행(날짜, 소계, 소계) 제거
df_filtered = df_filtered.drop(index='날짜')

# 변환된 데이터 확인
df_filtered
```
