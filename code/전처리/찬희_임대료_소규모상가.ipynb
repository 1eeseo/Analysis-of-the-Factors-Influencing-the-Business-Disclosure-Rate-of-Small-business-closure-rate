# 열 이름 변경

```python
def rename_columns(df):
    new_columns = []
    for col in df.columns:
        if "분기" in col:
            year, quarter = col.split("년")
            year = year.strip()[-2:]
            quarter = quarter.strip()[0]
            new_columns.append(f"{year}_{quarter}")
        else:
            new_columns.append(col)
    df.columns = new_columns
    return df

df1 = rename_columns(df1)

print("df1의 변경된 열 이름:\n", df1.columns)

```

# 서울 데이터 남기고 다 삭제

```python
import pandas as pd


def filter_seoul(df):
    filtered_df = df[df['지역'].str.startswith('서울')]
    return filtered_df


df1_filtered = filter_seoul(df1)
```
