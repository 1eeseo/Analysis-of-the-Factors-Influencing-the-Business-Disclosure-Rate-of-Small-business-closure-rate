```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "/content/drive/MyDrive/파이썬 시각화/data/1인당 국민총소득.xlsx"
df = pd.read_excel(path)
df.head()
```

```python
df.iloc[0, 0] = 'GNI'
```

```python
df.head()
```

```python
# GNI 데이터만 추출
gni_data = df.iloc[0, 1:].astype(float)  # GNI와 연도 데이터를 추출하여 숫자로 변환
years = df.columns[1:]  # 연도 추출

# 시각화
plt.figure(figsize=(12, 6))
sns.lineplot(x=years, y=gni_data, marker='o', color='b')

# 데이터 포인트에 값 표시
for i, value in enumerate(gni_data):
    if i % 5 == 0:
        plt.text(x=years[i], y=value, s=f'{value:.1f}',
                 horizontalalignment='center', verticalalignment='bottom', rotation=45)
        plt.title('GNI_Graph')
        plt.xlabel('Year')
        plt.ylabel('GNI')
        plt.grid()
```
