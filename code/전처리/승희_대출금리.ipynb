# 행, 열 바꾸기

```python
import pandas as pd

df = pd.read_csv(path1, index_col=0)

print("원본 데이터프레임:")
print(df)

df = df.T

print("\n행과 열이 변경된 데이터프레임:")
df
```

```python
# 인덱스와 컬럼을 결합하여 '연도월' 형태로 변환
df = df.stack()

# 인덱스 변환: '1월' -> '01', '2월' -> '02'
df.index = [f"{year}{month.replace('월', '').zfill(2)}" for month, year in df.index]

# 인덱스 정렬
df = df.reset_index()

# 자치구로 컬럼명 변경
df = df.rename(columns = {'index':'날짜'})
df.columns.values[1] = '대출금리'

df
```

```python
df.info()
```

```python
# '날짜' 컬럼을 datetime 형식으로 변환
df['날짜'] = pd.to_datetime(df['날짜'], format='%Y%m')

# 년도와 월 컬럼 생성
df['년도'] = df['날짜'].dt.year
df['월'] = df['날짜'].dt.month

# 2019-01 형태로 변환
df['날짜'] = df['날짜'].dt.strftime('%Y-%m')

df
```
